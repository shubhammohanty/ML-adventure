# KNN (K Nearest Neighbors Classifier)

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
data.feature_names
```

```python
data.target_names
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y0 = (df[df.target==0].shape)[0]
y1 = (df[df.target==1].shape)[0]
y2 = (df[df.target==2].shape)[0]
```

```python

plt.scatter(df["sepal length (cm)"][:y0], df["petal length (cm)"][:y0], color="red", marker = ".")
plt.scatter(df["sepal length (cm)"][y0:y0+y1], df["petal length (cm)"][y0:y0+y1], color="blue", marker = "*")
plt.scatter(df["sepal length (cm)"][y0+y1:y0+y1+y2], df["petal length (cm)"][y0+y1:y0+y1+y2], color="green", marker = "+")
```

```python

plt.scatter(df["sepal width (cm)"][:y0], df["petal width (cm)"][:y0], color="red", marker = ".")
plt.scatter(df["sepal width (cm)"][y0:y0+y1], df["petal width (cm)"][y0:y0+y1], color="blue", marker = "*")
plt.scatter(df["sepal width (cm)"][y0+y1:y0+y1+y2], df["petal width (cm)"][y0+y1:y0+y1+y2], color="green", marker = "+")
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']], df.target, train_size=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(xTrain, yTrain)
```

```python
model.score(xTest, yTest)
```
